```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
historical_table_html = open('historical_standings_html.txt', 'r')
soup = BeautifulSoup(historical_table_html, 'lxml')
historical_table_html.close()
```

```python
historical_east_table = soup.select('table.firstdiv')
historical_east_table_df = pd.DataFrame()
historical_east_table_df = pd.read_html(str(historical_east_table))[0]
# batting_stats_table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
historical_east_table_df
```

```python
historical_west_table = soup.select('table.seconddiv')
historical_west_table_df = pd.DataFrame()
historical_west_table_df = pd.read_html(str(historical_west_table))[0]
# batting_stats_table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
historical_west_table_df
```

```python
with pd.ExcelWriter("historical_standings.xlsx", mode="a", engine="openpyxl", if_sheet_exists="replace") as writer:
  historical_east_table_df.to_excel(writer, sheet_name="2012_regular_season", startrow=0, startcol=0)
  # historical_west_table_df.to_excel(writer, sheet_name="2015_regular_season", startrow=0, startcol=10)
```
