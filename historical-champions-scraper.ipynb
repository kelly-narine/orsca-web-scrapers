```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
champions_table_html = open('historical_champions_html.txt', 'r')
soup = BeautifulSoup(champions_table_html, 'lxml')
champions_table_html.close()
```

```python
champions_table = soup.select('table.champtbl')
champions_table_df = pd.DataFrame()
champions_table_df = pd.read_html(str(champions_table))[0]
# batting_stats_table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
champions_table_df
```

```python
with pd.ExcelWriter("historical_champions.xlsx", mode="a", engine="openpyxl", if_sheet_exists="replace") as writer:
  champions_table_df.to_excel(writer, sheet_name="champions", startrow=0, startcol=0, index=False)
  # bowling_stats_table_df.to_excel(writer, sheet_name="premier", startrow=14, startcol=0, index=False)
```
