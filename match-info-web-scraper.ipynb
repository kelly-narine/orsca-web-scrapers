Imports

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

Web Scraping

```python
url = 'https://orsca.ca/Pages/ArchiveStandings.aspx'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'lxml')

print(soup.prettify())
```

```python
table = soup.select('table.scoreboard')[3]
table_df = pd.read_html(str(table))[0].transpose()
table_df
```
