Imports

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

Web Scraping

```python
page1_html = open('page1_html.txt', 'r')
soup1 = BeautifulSoup(page1_html, 'lxml')
page1_html.close()

page2_html = open('page2_html.txt', 'r')
soup2 = BeautifulSoup(page2_html, 'lxml')
page2_html.close()

page3_html = open('page3_html.txt', 'r')
soup3 = BeautifulSoup(page3_html, 'lxml')
page3_html.close()
```

ETL (Extract, Transform, Load) Process

```python
table1 = soup1.select('table.scoreboard')
table2 = soup2.select('table.scoreboard')
table3 = soup3.select('table.scoreboard')
table_df = pd.DataFrame()

for match in table1:
  table_df = table_df.append(pd.read_html(str(match))[0].transpose())

for match in table2:
  table_df = table_df.append(pd.read_html(str(match))[0].transpose())

for match in table3:
  table_df = table_df.append(pd.read_html(str(match))[0].transpose())

table_df.columns = ['Match Date', 'Result', 'Team 1', 'Team 2']
table_df.to_csv('scoreboard_2022.csv', index=False)
table_df
```
