```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
points_table_html = open('points_table_regular_season_html.txt', 'r')
soup = BeautifulSoup(points_table_html, 'lxml')
points_table_html.close()
```

```python
elite_table = soup.select('table.firstdiv')
elite_table_df = pd.DataFrame()
elite_table_df = pd.read_html(str(elite_table))[0]
# table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
elite_table_df
```

```python
premier_table = soup.select('table.seconddiv')
premier_table_df = pd.DataFrame()
premier_table_df = pd.read_html(str(premier_table))[0]
# table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
premier_table_df
```

```python
with pd.ExcelWriter("points_table.xlsx", mode="a", engine="openpyxl", if_sheet_exists="replace") as writer:
  elite_table_df.to_excel(writer, sheet_name="regular_season", startrow=1, startcol=0, index=False)
  premier_table_df.to_excel(writer, sheet_name="regular_season", startrow=1, startcol=10, index=False)
```
