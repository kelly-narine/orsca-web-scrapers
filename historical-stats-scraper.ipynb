```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
historical_stats_table_html = open('historical_stats_html.txt', 'r')
soup = BeautifulSoup(historical_stats_table_html, 'lxml')
historical_stats_table_html.close()
```

```python
historical_batting_table = soup.select('table.ststats')[0]
historical_batting_table_df = pd.DataFrame()
historical_batting_table_df = pd.read_html(str(historical_batting_table))[0]
# batting_stats_table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
historical_batting_table_df
```

```python
historical_bowling_table = soup.select('table.ststats')[1]
historical_bowling_table_df = pd.DataFrame()
historical_bowling_table_df = pd.read_html(str(historical_bowling_table))[0]
# batting_stats_table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
historical_bowling_table_df
```

```python
with pd.ExcelWriter("historical_stats.xlsx", mode="a", engine="openpyxl", if_sheet_exists="replace") as writer:
  historical_batting_table_df.to_excel(writer, sheet_name="2012_regular_season", startrow=0, startcol=0)
  historical_bowling_table_df.to_excel(writer, sheet_name="2012_regular_season", startrow=231, startcol=0)
```
