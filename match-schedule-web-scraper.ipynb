```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
# !pip install pandasql
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
match_schedule_html = open('2022_t10_schedule_html.txt', 'r')
soup = BeautifulSoup(match_schedule_html, 'lxml')
match_schedule_html.close()
```

```python
table = soup.select('table.schedtbl')[0]
table_df = pd.DataFrame()
table_df = pd.read_html(str(table))[0]
# table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
table_df.to_csv('2022_t10_schedule.csv', index=False)
table_df
```
