```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
trophy_table_html = open('trophy_html.txt', 'r')
soup = BeautifulSoup(trophy_table_html, 'lxml')
trophy_table_html.close()
```

```python
trophy_table_100 = soup.select('table.trophytbl')[0]
trophy_table_100_df = pd.DataFrame()
trophy_table_100_df = pd.read_html(str(trophy_table_100))[0]
# batting_stats_table_df.drop(0, axis=0, inplace=True) # only drop first row if it contains default data
trophy_table_100_df
```
